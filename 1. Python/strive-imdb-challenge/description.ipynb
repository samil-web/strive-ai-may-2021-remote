```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get("https://www.imdb.com/list/ls066247137/")
print(response)
```

```python
page = response.content
soup = BeautifulSoup(page, "html.parser")
```

```python
descriptions = soup.find_all('p', class_ = "")[1::2]

# print(descriptions)

descriptions_list = []
for description in descriptions:
    descriptions_list.append(description.get_text().strip())

# print(descriptions_list)
for movie, description in enumerate(descriptions_list):
    print(f"\nDescription movie number {movie}: ", description)
# print(descriptions[1])
# descriptions[7].get_text().strip()
```
