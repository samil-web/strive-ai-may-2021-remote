<h3>1. Employment Type</h3>

```python
# Get rows where Employment Type has Nan value
empty_Employment = df[df['Employment.Type'].isnull()]['Employment.Type']
print(empty_Employment)
```

```python
# Find percentage of rows where Employment type is nan
Nan_percent = df['Employment.Type'].isnull().sum()*100/len(df)
Nan_percent
```

```python
df['Employment.Type'].unique()
```

If we want to play around with this column's data, we will need to change this to numerical values.<br><br> 
Like categories:<br>
1: 'Salary'<br>
2: 'Self employed'<br>
3: nan<br>
Can use the replace method for this.

```python
df['Employment.Type.Numeric'] = df['Employment.Type'].replace({'Salaried': 1, 'Self employed': 2, np.nan: 3})
```

```python
df.head()
```

```python
# Employment numeric type has no null values as they are replaced by 3.
# Can manipulate accordingly later
df.isnull().sum()
```

<h3>2: Date of Birth, 3: Disbursal Date</h3>

<ol>
    <li>https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.to_datetime.html</li>
    <li>https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_csv.html ---> date_parser </li>
</ol>

```python
df['Date.of.Birth.datetime'] = pd.to_datetime(df['Date.of.Birth'], format = '%d-%m-%y')
```

```python
df['DisbursalDate.datetime'] = df['DisbursalDate'].apply(lambda x: datetime.strptime(x, '%d-%m-%y'))
```

```python
df.head()
```

```python
# Check null values to make sure all dates were parsed and formatted
df.isnull().sum()
```

```python
# https://stackoverflow.com/questions/29206612/difference-between-data-type-datetime64ns-and-m8ns
df['Date.of.Birth.datetime'].dtype
# np.dtype('datetime64[ns]') == np.dtype('<M8[ns]')
```

```python
# Extracting only the rows where date time parsing gives year > 2021
df[df['Date.of.Birth.datetime'].dt.year > 2021]['Date.of.Birth']
```

```python
df[df['DisbursalDate.datetime'].dt.year > 2021]['DisbursalDate']
```

<h4>Explanation for above behavior</h4><br>
<a href = 'https://stackoverflow.com/questions/37766353/pandas-to-datetime-parsing-wrong-year'>Cutoff year</a> in python datetime library
<p>Below '31-Dec-68' --> 2068 <br>
   Above '1-Jan-69' --> 1968
</p>
